```python
# Defining the Menu for restaurant

menu = { "Samosa" : 20,
         "Kachori": 15,
         "Dhokla": 15,
         "Jalebi": 10,
         "Panipuri": 30
         }

#create a placeholder for order & bill - to add items in cart

order = {}
bill = 0

#display welcome message & show menu - I used double line for visuallly appealing menu.

print("\n====================================================")
print(f"Welcome to Govinda's Restaurant..! Here is our Menu")
print("====================================================")
print('''
        ☆ Samosa    :  ₹ 20
        ☆ Kachori   :  ₹ 15
        ☆ Dhokla    :  ₹ 15
        ☆ Jalebi    :  ₹ 10
        ☆ Panipuri  :  ₹ 30
''')
print("====================================================\n")

# use done as a flag if done is false - the while loop condition will run as long as while loop is true.

Done = False

while not Done:
    flag = 0

    #ask customer to input their order - convert that order.

    item = input(f"What you would like to order? [Enter 'Done' to finish the Order]: \n").title()

    #if item present in the menu then only ask customer for quantity.

    if item in menu:
        qty = input(f"How many {item}, you would like to order?\n")

        # to protect qty from string input, we used isnumeric function 

        if qty.isnumeric() == True:

            # if item is already exists in order, then add qty in existing qty.
            if qty in order:
                order[item] = order[item] + int(qty)

            # if item is already not exists in order then add it in order 

            else:
                order[item] = int(qty)
          #if qty is in negative or string format --> then ask customer to provide input in positive integer.
        else:
            print(f"Invalid Input! Please Enter the Positive Number:\n")

    #if user said Done then check if the user is leaving hotel without any order ..?
    elif item == 'Done':

        # if the size of customer order is equal to zero. it means custormer didn't order anything. 
        if len(order) == 0:
            #ask customer to order something.
            print('No Ordered Place!, Kindly place the Order.\n')
            # we asked customer for confirmation, they really want to leave the restaurant without ordering the food..?
            choice = input(f"Please Enter 'Y' or 'N'. Press 'Y' to continue the order or press 'N' to exit: \n").upper()

            #will take action as per the cutomer reply.
            if choice == 'Y':
                flag = 0
            elif choice == 'N':
                print(f"Thanks for Visiting our Restauarant. Have a wonderful Day!")
                Done = True
                break
            else:
                print(f"Invalid Input. Please Enter 'Y' or 'N'. Press 'Y' to continue the order or press 'N' to exit: \n")
        else:
            print(f'Thank you for placing your order, we appreciate your choice.\n')
            Done = True
    #sometimes customer will order that food that is not available in menu 
    else:
        print(f"Sorry!, {item} is not available. Please choose any Dish from above given Menu. Thank you!\n")
print("===============================================")
print("  *** Here is your Order Summary ***     ")
print("===============================================")

#creating a bill receipt for customer.

for item, qty in order.items():

    #calculate the price of each item  & then add that item, then add all item bill amount in total bill
    item_cost = menu[item] * int(qty)
    print(f"        {item} : {int(qty)} X {menu[item]} = {item_cost}")
    bill = item_cost + bill
print("===============================================")
print(f"        You're Total Bill is ₹{bill}")
print("===============================================")
print("        Thank you, and have a great day!")

```
