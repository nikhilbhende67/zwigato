```python
# Defining the Menu for restaurant

menu = { "Samosa" : 20,
         "Kachori": 15,
         "Dhokla": 15,
         "Jalebi": 10,
         "Panipuri": 30
         }

#create a placeholder for order - to add items in cart

order = {}
bill = 0

print("\n====================================================")
print(f"Welcome to Govinda's Restaurant..! Here is our Menu")
print("====================================================")
print('''
        ☆ Samosa    :  ₹ 20
        ☆ Kachori   :  ₹ 15
        ☆ Dhokla    :  ₹ 15
        ☆ Jalebi    :  ₹ 10
        ☆ Panipuri  :  ₹ 30
''')
print("====================================================\n")
Done = False

while not Done:
    flag = 0
    item = input(f"What you would like to order? [Enter 'Done' to finish the Order]: \n").title()
    if item in menu:
        qty = input(f"How many {item}, you would like to order?\n")
        if qty.isnumeric() == True:
            if qty in order:
                order[item] = order[item] + int(qty)
            else:
                order[item] = int(qty)
            order[item] = int(qty)
        else:
            print(f"Invalid Input! Please Enter the Positive Number:\n")
    elif item == 'Done':
        if len(order) == 0:
            print('No Ordered Place!, Kindly place the Order.\n')
            choice = input(f"Please Enter 'Y' or 'N'. Press 'Y' to continue the order or press 'N' to exit: \n").upper()
            if choice == 'Y':
                flag = 0
            elif choice == 'N':
                print(f"Thanks for Visiting our Restauarant. Have a wonderful Day!")
                Done = True
                break
            else:
                print(f"Invalid Input. Please Enter 'Y' or 'N'. Press 'Y' to continue the order or press 'N' to exit: \n")
        else:
            print(f'Thank you for placing your order, we appreciate your choice.')
            Done = True
    else:
        print(f"Sorry!, {item} is not available. Please choose any Dish from above given Menu. Thank you!\n")
print("===============================================")
print("  *** Here is your Order Summary ***     ")
print("===============================================")

for item, qty in order.items():
    item_cost = menu[item] * int(qty)
    print(f"        {item} : {int(qty)} X {menu[item]} = {item_cost}")
    bill = item_cost + bill
print("===============================================")
print(f"         You're Total Bill is ₹{bill}")
print("===============================================\n")
print("         Thank you, and have a great day!")

```
